```python
# ==============================
# CRICKET PLAYER PERFORMANCE SYSTEM
# ==============================

# Base class
class Person:
    def __init__(self, name, team_name):
        self.name = name
        self.team_name = team_name

    def display_name(self):
        print("Player Name        :", self.name)
        print("Playing For Team   :", self.team_name)


# Derived class
class Player(Person):
    def __init__(self, name, team_name, runs, overs):
        super().__init__(name, team_name)
        self.runs = runs
        self.overs = overs
        self.balls = self.convert_overs_to_balls()

    # Convert overs into total balls
    def convert_overs_to_balls(self):
        whole_overs = int(self.overs)
        extra_balls = round((self.overs - whole_overs) * 10)

        if extra_balls > 5:
            return 0

        return (whole_overs * 6) + extra_balls

    # Calculate strike rate
    def calculate_strike_rate(self):
        if self.balls == 0:
            return 0
        return (self.runs / self.balls) * 100

    # Performance category
    def performance_category(self):
        sr = self.calculate_strike_rate()
        if sr < 100:
            return "Poor"
        elif sr < 120:
            return "Average"
        elif sr < 150:
            return "Good"
        else:
            return "Excellent"

    # Display individual player details
    def display(self):
        print("\n--------------------------------")
        print("        PLAYER SUMMARY")
        print("--------------------------------")
        self.display_name()
        print("Runs Scored        :", self.runs)
        print("Overs Played       :", self.overs)
        print("Balls Faced        :", self.balls)
        print("Strike Rate        : {:.2f}".format(self.calculate_strike_rate()))
        print("Performance Level  :", self.performance_category())
        print("--------------------------------")

    # Return summary for file saving
    def get_summary(self):
        summary = "\n--------------------------------\n"
        summary += "        PLAYER SUMMARY\n"
        summary += "--------------------------------\n"
        summary += f"Player Name        : {self.name}\n"
        summary += f"Playing For Team   : {self.team_name}\n"
        summary += f"Runs Scored        : {self.runs}\n"
        summary += f"Overs Played       : {self.overs}\n"
        summary += f"Balls Faced        : {self.balls}\n"
        summary += "Strike Rate        : {:.2f}\n".format(self.calculate_strike_rate())
        summary += f"Performance Level  : {self.performance_category()}\n"
        summary += "--------------------------------\n"
        return summary


# ==============================
# FUNCTION TO SHOW TEAM LIST
# ==============================
def show_teams(team_list):
    print("\nAvailable Teams:")
    for i, team in enumerate(team_list, start=1):
        print(f"{i}. {team}")


# ==============================
# FUNCTION TO SHOW PLAYER LIST
# ==============================
def show_players(players):
    print("\n==========================================")
    print("             PLAYER LIST")
    print("==========================================")

    if len(players) == 0:
        print("No players added yet.")
    else:
        for i, p in enumerate(players, start=1):
            print(f"{i}. {p.name} | Runs: {p.runs} | Overs: {p.overs} | Balls: {p.balls}")

    print("==========================================")


# ==============================
# FUNCTION TO TAKE PLAYER INPUT
# ==============================
def input_player(team_name):
    name = input("Enter player name: ").strip()

    while name == "":
        print("Player name cannot be empty.")
        name = input("Enter player name: ").strip()

    # Runs input
    while True:
        try:
            runs = int(input("Enter runs scored: "))
            if runs < 0:
                print("Runs cannot be negative.")
                continue
            break
        except ValueError:
            print("Invalid input! Please enter an integer for runs.")

    # Overs input
    while True:
        try:
            overs = float(input("Enter overs played: "))
            whole = int(overs)
            extra = round((overs - whole) * 10)

            if overs < 0:
                print("Overs cannot be negative.")
                continue

            if extra > 5:
                print("Invalid overs! After decimal only 0 to 5 allowed.")
                continue
            break
        except ValueError:
            print("Invalid input! Please enter overs like 4.2 or 3.5")

    return Player(name, team_name, runs, overs)


# ==============================
# FUNCTION TO DISPLAY FINAL RESULT
# ==============================
def display_results(players_list, tournament_name, team_name):
    # DISPLAY ALL PLAYERS
    print("\n==========================================")
    print("         ALL PLAYERS SUMMARY")
    print("==========================================")
    print("Tournament Name     :", tournament_name)
    print("Team Name           :", team_name)

    for p in players_list:
        p.display()

    # TEAM ANALYSIS
    total_runs = sum(p.runs for p in players_list)
    total_balls = sum(p.balls for p in players_list)
    average_sr = sum(p.calculate_strike_rate() for p in players_list) / len(players_list)

    highest_scorer = max(players_list, key=lambda p: p.runs)
    best_sr_player = max(players_list, key=lambda p: p.calculate_strike_rate())

    team_overs = f"{total_balls // 6}.{total_balls % 6}"

    print("\n==========================================")
    print("            TEAM ANALYSIS")
    print("==========================================")
    print("Player Name        :", highest_scorer.name)
    print("Tournament Name    :", tournament_name)
    print("Team Name          :", team_name)
    print("Total Team Runs    :", total_runs)
    print("Total Team Balls   :", total_balls)
    print("Total Team Overs   :", team_overs)
    print("Average Strike Rate: {:.2f}".format(average_sr))
    print("Highest Score      :", highest_scorer.runs, "runs")
    print("Best Strike Rate   : {:.2f}".format(best_sr_player.calculate_strike_rate()))
    print("==========================================")


# ==============================
# FUNCTION TO SAVE TO FILE
# ==============================
def save_to_file(players_list, tournament_name, team_name):
    total_runs = sum(p.runs for p in players_list)
    total_balls = sum(p.balls for p in players_list)
    average_sr = sum(p.calculate_strike_rate() for p in players_list) / len(players_list)

    highest_scorer = max(players_list, key=lambda p: p.runs)
    best_sr_player = max(players_list, key=lambda p: p.calculate_strike_rate())

    team_overs = f"{total_balls // 6}.{total_balls % 6}"

    file = open("cricket.txt", "w")

    file.write("==========================================\n")
    file.write("   CRICKET PLAYER PERFORMANCE SYSTEM\n")
    file.write("==========================================\n")
    file.write(f"Tournament Name     : {tournament_name}\n")
    file.write(f"Team Name           : {team_name}\n")

    for p in players_list:
        file.write(p.get_summary())

    file.write("\n==========================================\n")
    file.write("            TEAM ANALYSIS\n")
    file.write("==========================================\n")
    file.write(f"Player Name         : {highest_scorer.name}\n")
    file.write(f"Tournament Name     : {tournament_name}\n")
    file.write(f"Team Name           : {team_name}\n")
    file.write(f"Total Team Runs     : {total_runs}\n")
    file.write(f"Total Team Balls    : {total_balls}\n")
    file.write(f"Total Team Overs    : {team_overs}\n")
    file.write("Average Strike Rate : {:.2f}\n".format(average_sr))
    file.write(f"Highest Score       : {highest_scorer.runs} runs\n")
    file.write("Best Strike Rate    : {:.2f}\n".format(best_sr_player.calculate_strike_rate()))
    file.write("==========================================\n")

    file.close()


# ==============================
# MAIN PROGRAM
# ==============================
print("==========================================")
print("   CRICKET PLAYER PERFORMANCE SYSTEM")
print("==========================================")

# ------------------------------
# Tournament Selection
# ------------------------------
print("\nSelect Tournament Type:")
print("1. IPL")
print("2. World Cup")

while True:
    try:
        tournament_choice = int(input("Enter your choice (1 or 2): "))
        if tournament_choice not in [1, 2]:
            print("Please select only 1 or 2.")
            continue
        break
    except ValueError:
        print("Invalid input! Please enter 1 or 2.")

# ------------------------------
# Team Lists
# ------------------------------
ipl_teams = [
    "Chennai Super Kings",
    "Mumbai Indians",
    "Royal Challengers Bengaluru",
    "Kolkata Knight Riders",
    "Rajasthan Royals",
    "Sunrisers Hyderabad",
    "Delhi Capitals",
    "Punjab Kings",
    "Lucknow Super Giants",
    "Gujarat Titans"
]

worldcup_teams = [
    "India",
    "Australia",
    "England",
    "Pakistan",
    "South Africa",
    "New Zealand",
    "Sri Lanka",
    "Bangladesh",
    "Afghanistan",
    "West Indies"
]

if tournament_choice == 1:
    tournament_name = "IPL"
    selected_team_list = ipl_teams
else:
    tournament_name = "World Cup"
    selected_team_list = worldcup_teams

# ------------------------------
# TEAM MANAGEMENT MENU
# ------------------------------
while True:
    print(f"\n===== {tournament_name} TEAM MANAGEMENT =====")
    print("1. Show Teams")
    print("2. Add New Team")
    print("3. Remove Team")
    print("4. Continue with Team Selection")

    try:
        manage_choice = int(input("Enter your choice: "))

        if manage_choice == 1:
            show_teams(selected_team_list)

        elif manage_choice == 2:
            new_team = input("Enter new team name: ").strip()
            if new_team == "":
                print("Team name cannot be empty.")
            elif new_team in selected_team_list:
                print("Team already exists.")
            else:
                selected_team_list.append(new_team)
                print(f"✅ Team '{new_team}' added successfully.")

        elif manage_choice == 3:
            show_teams(selected_team_list)
            try:
                remove_index = int(input("Enter team number to remove: "))
                if 1 <= remove_index <= len(selected_team_list):
                    removed_team = selected_team_list.pop(remove_index - 1)
                    print(f"❌ Team '{removed_team}' removed successfully.")
                else:
                    print("Invalid team number.")
            except ValueError:
                print("Please enter a valid number.")

        elif manage_choice == 4:
            break

        else:
            print("Please choose a valid option (1 to 4).")

    except ValueError:
        print("Invalid input! Please enter a number.")

# ------------------------------
# Final Team Selection
# ------------------------------
print(f"\nSelect {tournament_name} Team:")
show_teams(selected_team_list)

while True:
    try:
        team_choice = int(input("Enter team number: "))
        if 1 <= team_choice <= len(selected_team_list):
            team_name = selected_team_list[team_choice - 1]
            break
        else:
            print("Please select a valid team number.")
    except ValueError:
        print("Invalid input! Please enter a valid number.")

# ------------------------------
# Input number of players
# ------------------------------
while True:
    try:
        num_players = int(input("\nEnter number of players: "))
        if num_players <= 0:
            print("Please enter a positive number.")
            continue
        break
    except ValueError:
        print("Invalid input! Please enter a valid integer.")

players_list = []

# ------------------------------
# Initial Player Input
# ------------------------------
for i in range(num_players):
    print(f"\nEnter details for Player {i+1}")
    print("--------------------------------")
    player = input_player(team_name)
    players_list.append(player)

# ------------------------------
# PLAYER MANAGEMENT MENU
# ------------------------------
while True:
    print("\n===== PLAYER MANAGEMENT =====")
    print("1. Show Players")
    print("2. Add New Player")
    print("3. Remove Player")
    print("4. Continue to Result")

    try:
        player_choice = int(input("Enter your choice: "))

        if player_choice == 1:
            show_players(players_list)

        elif player_choice == 2:
            print("\nEnter details for New Player")
            print("--------------------------------")
            new_player = input_player(team_name)
            players_list.append(new_player)
            print(f"✅ Player '{new_player.name}' added successfully.")

        elif player_choice == 3:
            if len(players_list) == 0:
                print("No players available to remove.")
            else:
                show_players(players_list)
                try:
                    remove_index = int(input("Enter player number to remove: "))
                    if 1 <= remove_index <= len(players_list):
                        removed_player = players_list.pop(remove_index - 1)
                        print(f"❌ Player '{removed_player.name}' removed successfully.")
                    else:
                        print("Invalid player number.")
                except ValueError:
                    print("Please enter a valid number.")

        elif player_choice == 4:
            if len(players_list) == 0:
                print("At least one player is required.")
            else:
                break

        else:
            print("Please choose a valid option (1 to 4).")

    except ValueError:
        print("Invalid input! Please enter a number.")

# ------------------------------
# RESULT + ASK AGAIN TO ADD MORE PLAYER
# ------------------------------
while True:
    display_results(players_list, tournament_name, team_name)
    save_to_file(players_list, tournament_name, team_name)

    print("\nDo you want to add more player?")
    print("1. Yes")
    print("2. No / Exit")

    try:
        more_choice = int(input("Enter your choice: "))

        if more_choice == 1:
            print("\nEnter details for New Player")
            print("--------------------------------")
            new_player = input_player(team_name)
            players_list.append(new_player)
            print(f"✅ Player '{new_player.name}' added successfully.")
            print("\n🔄 Updated Result Generated Successfully!")

        elif more_choice == 2:
            print("\n🏏 Thank you for using Cricket Player Performance System!")
            print("Program Ended Successfully.")
            break

        else:
            print("Please choose only 1 or 2.")

    except ValueError:
        print("Invalid input! Please enter 1 or 2.")
```
